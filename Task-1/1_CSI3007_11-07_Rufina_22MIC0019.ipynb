```python
import pdfplumber
from PIL import Image
import pytesseract
import io

# Set tesseract path
pytesseract.pytesseract.tesseract_cmd = r"D:\7TH SEMESTER\Adv_Python\Python_lab\Tesseract-OCR\tesseract.exe"

pdf_path = "sample_22mic0019.pdf"

# Open the PDF
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        print(f"\n--- Page {i+1} ---")
        # Get all images in the page
        images = page.images

        if not images:
            print("No images found.")
        else:
            for j, img in enumerate(images):
                # Extract image object
                bbox = (img["x0"], img["top"], img["x1"], img["bottom"])
                cropped = page.crop(bbox)
                img_obj = cropped.to_image(resolution=300)

                # OCR on extracted image
                text = pytesseract.image_to_string(img_obj.original)
                print(f"Text from image {j+1}:\n{text}")
```

```python
import re  # Import the regular expressions module for pattern matching and text cleaning

def clean_ocr_text(text):
    # Step 1: Fix common misreads
    text = re.sub(r'emp\s+loyee', 'employee', text, flags=re.I)
    text = re.sub(r'printC', 'print(', text)
    text = re.sub(r'catC', 'cat(', text)
    text = re.sub(r'summary\s*\(emp\s*\|?loyee.*?\)', 'summary(employee_df)', text, flags=re.I)
    text = re.sub(r'cC"', 'c("', text)
    text = re.sub(r'“|”', '"', text)
    text = re.sub(r'\|', '', text)  # Remove vertical bars
    text = re.sub(r'[\u2022\u00a0]', ' ', text)  # Remove bullet or non-breaking space
    text = re.sub(r' +', ' ', text)  # Reduce multiple spaces

    # Step 2: Remove random garbage lines (short non-code lines)
    lines = text.split('\n')
    cleaned_lines = []
    for line in lines:
        line = line.strip()
        if not line:
            continue
        # Remove lines with only numbers, symbols, or very short noise
        if re.match(r'^[^a-zA-Z0-9]{1,3}$', line) or re.match(r'^[0-9]+$', line):
            continue
        if len(line) < 4 and not re.search(r'[a-zA-Z]', line):
            continue
        cleaned_lines.append(line)

    # Step 3: Join and normalize spacing
    return '\n'.join(cleaned_lines)
```

```python
import pdfplumber
from PIL import Image
import pytesseract
import io

# Set tesseract path
pytesseract.pytesseract.tesseract_cmd = r"D:\7TH SEMESTER\Adv_Python\Python_lab\Tesseract-OCR\tesseract.exe"

pdf_path = "sample_22mic0019.pdf"

# Open the PDF
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        print(f"\n--- Page {i+1} ---")
        # Get all images in the page
        images = page.images

        if not images:
            print("No images found.")
        else:
            for j, img in enumerate(images):
                # Extract image object
                bbox = (img["x0"], img["top"], img["x1"], img["bottom"])
                cropped = page.crop(bbox)
                img_obj = cropped.to_image(resolution=300)

                # OCR on extracted image
                raw_text = pytesseract.image_to_string(img_obj.original)

                #Removing noise from the text
                cleaned_text = clean_ocr_text(raw_text)
                
                print(f"Cleaned Text from image {j+1}:\n{cleaned_text}")

```

```python
import pdfplumber
from PIL import Image
import pytesseract
import io

# Set tesseract path
pytesseract.pytesseract.tesseract_cmd = r"D:\7TH SEMESTER\Adv_Python\Python_lab\Tesseract-OCR\tesseract.exe"

pdf_path = "sample_22mic0019.pdf"

# Open the PDF
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        print(f"\n--- Page {i+1} ---")
        # Get all images in the page
        images = page.images

        if not images:
            print("No images found.")
        else:
            for j, img in enumerate(images):
                # Extract image object
                bbox = (img["x0"], img["top"], img["x1"], img["bottom"])
                cropped = page.crop(bbox)
                img_obj = cropped.to_image(resolution=300)

                # Set custom configuration for Tesseract OCR:
                # --oem 3 lets Tesseract choose the most effective engine, often the LSTM model or LSTM + legacy + combined
                # --psm 6 treats the image as a single uniform block of text (like a paragraph or table row)
                custom_config = r'--oem 3 --psm 6'
                # OCR on extracted image
                raw_text = pytesseract.image_to_string(img_obj.original, config=custom_config)

                #Removing noise from the text
                cleaned_text = clean_ocr_text(raw_text)
                
                print(f"Cleaned Text from image {j+1}:\n{cleaned_text}")

```

```python
'''If there is a proper table structure in the pdf, 
we can extract the table-content row wise and store each record in a list.'''

import pdfplumber

# Path to the sample PDF
pdf_path = "sample.pdf"

# Open the PDF
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        print(f"\n--- Page {i+1} ---\n")

        # 1. Extract and print plain text
        text = page.extract_text()
        print("Extracted Text:\n", text)

        # 2. Extract and print tables (if any)
        tables = page.extract_tables()
        if tables:
            for idx, table in enumerate(tables):
                print(f"\n Table {idx + 1}:")
                for row in table:
                    print(row)
        else:
            print("No tables found on this page.")
```
