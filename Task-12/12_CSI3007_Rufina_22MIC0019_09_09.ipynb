Working with CSV file.
   
   Reading csv file content as DataFrame
   
   Append data to csv
   
   Reading it chunk by chunk
   
       Based on chunk size
       Based on attribute filtering
   
   Write numeric data into the csv file
   
   Write text data to the csv file

```python
import pandas as pd
```

```python
file_path = "ncr_ride_bookings.csv"
df = pd.read_csv(file_path)
```

```python
print(df.head())
```

```python
new_df = pd.DataFrame([{"Date": "29-11-2024",	"Time": "18:01:39",	"Booking Id": "CNR1326809"}])
new_df.to_csv(file_path, mode = "a", header = False, index=False)
```

```python
size = 100

for chunk in pd.read_csv(file_path, chunksize=size):
    print(chunk.head())
```

```python
for vehicle, chunk in df.groupby("Vehicle Type"):
    print(f"Chunk for {vehicle}")
    print(chunk.head())
```

```python
numeric_df = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4.5, 5.6, 7.8]})
numeric_df.to_csv(file_path, mode='a', header=False, index=False)
print("Numeric data appended to CSV.")

# Read the file again to confirm data is appended
df_after_append = pd.read_csv(file_path)
print("\nContent of the CSV after appending:")
print(df_after_append.tail())
```

```python
text_df = pd.DataFrame({'colA': ['text1', 'text2', 'text3'], 'colB': ['text4', 'text5', 'text6']})
text_df.to_csv(file_path, mode='a', header=False, index=False)
print("Text data appended to CSV.")

# Read the file again to confirm data is appended
df_after_append_text = pd.read_csv(file_path)
print("\nContent of the CSV after appending text data:")
print(df_after_append_text.tail())
```
