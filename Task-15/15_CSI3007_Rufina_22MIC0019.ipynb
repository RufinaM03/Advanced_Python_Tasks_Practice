```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, filters, exposure
from scipy.ndimage import gaussian_filter, convolve
```

```python
img = io.imread("sparrow_img.jpg")
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
```

```python
plt.figure(figsize=(8,6))
plt.hist(img.ravel(), bins=255)
plt.title("Histogram of Image Intensities")
plt.show()
```

```python
gray = color.rgb2gray(img)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

```python
bright = np.clip(img + 40, 0, 255).astype(np.uint8)
plt.imshow(bright)
plt.title("Brightness Increased")
plt.axis("off")
```

```python
dark = np.clip(img - 40, 0, 255).astype(np.uint8)
plt.imshow(dark)
plt.title("Brightness Decreased")
plt.axis("off")
```

```python
contrast_high = exposure.rescale_intensity(img, in_range=(50, 200))
plt.imshow(contrast_high)
plt.title("High Contrast")
plt.axis("off")
```

```python
contrast_low = exposure.rescale_intensity(img, in_range=(100, 150))
plt.imshow(contrast_low)
plt.title("Low Contrast")
plt.axis("off")
```

```python
negative = 255 - img
plt.imshow(negative)
plt.title("Negative Image")
plt.axis("off")
```

```python
threshold = gray > 0.5
plt.imshow(threshold, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")
```

```python
blur = gaussian_filter(img, sigma=3)
plt.imshow(blur)
plt.title("Gaussian Blur")
plt.axis("off")
```

```python
edges = filters.sobel(gray)
plt.imshow(edges, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis("off")
```

```python
small = img[::2, ::2]   # simple downscale
plt.imshow(small)
plt.title("Resized 50%")
plt.axis("off")
```

```python
crop = img[50:300, 100:400]
plt.imshow(crop)
plt.title("Cropped Image")
plt.axis("off")
```

```python
rotated = np.rot90(img)
plt.imshow(rotated)
plt.title("Rotated 90")
plt.axis("off")
```
