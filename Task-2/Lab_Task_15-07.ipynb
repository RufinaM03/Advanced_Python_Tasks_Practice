```python
from nltk.tokenize import word_tokenize, sent_tokenize
from urllib import request

url = "https://www.gutenberg.org/cache/epub/76503/pg76503-images.html"

response = request.urlopen(url)
raw = response.read().decode('utf8')
```

```python
import sys  
!{sys.executable} -m pip install html2text
```

```python
import html2text

#Converting html to text
text = html2text.html2text(raw)

print("Text: " , text)
```

```python
# Tokenize the text
tokens = word_tokenize(text)

# Find length of tokens
print("Number of tokens:", len(tokens))
```

```python
print(tokens[:500])
```

```python
# Part-of-speech tagging
tagged = nltk.pos_tag(tokens)
print("POS Tagged:", tagged[:20])
```

```python
#String Basics, String Manipulation

sample = "  \tHello\tWorld!  \nThis is Python.  "

print("Original:", repr(sample))

# Whitespace removal
print("strip():", sample.strip())           # Removes leading and trailing whitespace
print("lstrip():", sample.lstrip())         # Removes leading whitespace
print("rstrip():", sample.rstrip())         # Removes trailing whitespace

# Case conversion
print("lower():", sample.lower())           # Converts to lowercase
print("upper():", sample.upper())           # Converts to uppercase
print("title():", sample.title())           # Title case: First letter capital of each word
print("capitalize():", sample.capitalize()) # First letter capital, rest lowercase
print("swapcase():", sample.swapcase())     # Swap upper to lower and vice versa

# Search and replace
print("replace():", sample.replace("Python", "NLP"))  # Replace substring
print("startswith('  '):", sample.startswith("  "))   # Check if starts with specific substring
print("endswith('.'):", sample.endswith("."))         # Check if ends with specific substring
print("find('World'):", sample.find("World"))         # Returns index or -1 if not found
print("count('l'):", sample.count("l"))               # Count occurrences

# Splitting and joining
print("split():", sample.split())                     # Split on whitespace
print("splitlines():", sample.splitlines())           # Split at line breaks
print("join():", " ".join(["Hello", "there"]))        # Join list into string with spaces

# Alignment
print("center():", sample.center(40, '-'))            # Center align with fill
print("ljust():", sample.ljust(30, '_'))              # Left align with fill
print("rjust():", sample.rjust(30, '_'))              # Right align with fill
print("zfill():", "42".zfill(5))                      # Pad numeric string with zeros

# Character checks (returns True/False)
print("isalnum():", "Python3".isalnum())              # Only letters and numbers
print("isalpha():", "Hello".isalpha())                # Only letters
print("isdigit():", "12345".isdigit())                # Only digits
print("isspace():", "   \n\t".isspace())              # Only whitespace
print("islower():", "hello".islower())                # All lowercase
print("isupper():", "HELLO".isupper())                # All uppercase
```

```python
'''Multiline 
String Input'''

print("Enter multiline text (end with an empty line):")
lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)

multi_text = "\n".join(lines)
print("Multiline input:\n", multi_text)
```

```python
#To remove utf encoded characters
raw_text = "✅ This is a test\u2028 with strange\u200b characters \ufeff.❗"

# Remove Byte Order Mark (BOM) and non-ASCII characters
clean_text = raw_text.encode("ascii", "ignore").decode()
print("Cleaned text:", clean_text)
```

```python
'''How the above code works:
"ascii":
Means we want to encode the text to ASCII (which only supports characters from 0–127).

"ignore":
Tells Python to ignore any character that cannot be represented in ASCII (like emojis, special symbols, etc.).'''
```

```python
import spacy

# Load Chinese tokenizer mode
nlp = spacy.load("zh_core_web_sm")

text = "今天天气不错，我们去公园吧！"
doc = nlp(text)

tokens = [token.text for token in doc]
print("Chinese Tokens:", tokens)
```

```python
import spacy

# Load Japanese tokenizer model
nlp = spacy.load("ja_core_news_sm")

text = "今日は天気がいいですね。公園に行きましょう！"
doc = nlp(text)

tokens = [token.text for token in doc]
print("Japanese Tokens:", tokens)
```
